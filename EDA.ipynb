```python
quality_counts = df['quality'].value_counts()

plt.figure(figsize=(8,6))

plt.bar(
    quality_counts.index,
    quality_counts
)

plt.title('Count Plot of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.swarmplot(x="quality", y="alcohol", data=df)

plt.title('Swarm Plot for Wine Quality and Alcohol Content')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
df['quality'] = df['quality'].astype(str)

plt.figure(figsize=(10, 8))

sns.violinplot(x="quality", y="alcohol", data=df)

plt.title('Violin Plot for Wine Quality and Alcohol Content')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
sns.pairplot(df)

plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```
