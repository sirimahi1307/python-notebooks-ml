```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df=pd.read_csv('./WineQT.csv')
print(df.head())
```

```python
marks_counts = df['Marks'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(marks_counts.index, marks_counts)
plt.title('Count Plot of Marks')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.swarmplot(x="Marks", y="Age", data=df)

plt.title('Swarm Plot for Marks and Age')
plt.xlabel('Marks')
plt.ylabel('Age')
plt.show()
```

```python
df['Marks'] = df['Marks'].astype(str)

plt.figure(figsize=(10, 8))

sns.violinplot(x="Marks", y="Age", data=df)

plt.title('Violin Plot for Marks and Age')
plt.xlabel('Marks')
plt.ylabel('Age')
plt.show()
```

```python
sns.pairplot(df)

plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```
