#phase 3:visual and statistical analysis

```python
df.hist(figsize=(8,5))
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.show()
```

```python
sns.scatterplot(x='Age',y='Marks',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4)) 
sns.heatmap(df.corr(numeric_only=True), annot=True, )
plt.show()
```
