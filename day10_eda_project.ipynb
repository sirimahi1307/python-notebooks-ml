phase 2

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Age'] = pd.to_numeric(
    df['Age'],
    errors='coerce'
)

df['Marks'] = pd.to_numeric(
    df['Marks'],
    errors='coerce'
)
```

```python
df.isnull().sum()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.isnull().sum()
```

phase 3

```python
df.hist(figsize=(10,5))
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.show()
```

```python
sns.scatterplot(x='Age',y='Marks',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4)) 
sns.heatmap(df.corr(numeric_only=True), annot=True, )
plt.show()
```
