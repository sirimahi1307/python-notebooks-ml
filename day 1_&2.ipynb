```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Array:\n", arr)
print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
```

```python
a = np.array([1, 2, 3])
b = np.zeros((2, 2))
c = np.ones((2, 2))
d = np.arange(0, 10, 2)
e = np.empty((2, 2), dtype=int)

print("Matrix A \n", a, "\n Matrix B \n", b, "\n Matrix C \n", c, "\n Matrix D \n", d, "\n Matrix E \n", e)
```

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

```python
arr = np.array([10, 20, 30, 40])

print("First element:", arr[0])
print("Last element:", arr[-1])
```

```python
arr = np.array([10, 20, 30, 40, 50])

print("Slice (1 to 3):", arr[1:4])
print("First three elements:", arr[:3])
```

```python
arr = np.array([1, 2, 3])

print("Add scalar:", arr + 5)

b = np.array([[1], [2], [3]])
c = np.array([10, 20, 30])

print("Broadcasting result:\n", b + c)
```

```python
import pandas as pd

data = [10, 20, 30, 40]
series = pd.Series(data)

print(series)
print("First element:", series[0])
```

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "A", "B"],
    "Age": [20, 21, 22, 20, 21]
}
df = pd.DataFrame(data)

print("Info:")
print(df.info())

print("\nDescribe:\n", df.describe())

print("\nValue Counts:\n", df["Name"].value_counts())

print("\nHead:\n", df.head())

print("\nTail:\n", df.tail())
```
